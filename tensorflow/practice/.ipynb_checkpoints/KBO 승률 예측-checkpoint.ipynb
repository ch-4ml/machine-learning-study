### 학습 데이터와 테스트 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=64)
```

### 스케일링

```python
# 정규화(= StandardScaler)
X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
X_train_std = (X_train - X_mean) / X_std
X_test_std = (X_test - X_mean) / X_std
```

```python
X_train_std.head()
```

## 회귀 분석

먼저 기계학습을 통해 회귀 분석을 진행한다.

남은 특성들도 일부 상관관계가 있기 때문에 규제가 있는 모델을 이용해서 학습한다.

### 모델 구성

중요하지 않거나, 상관관계가 높은 Feature에 규제를 가하기 위해서 <b>규제가 있는 모델인 Ridge, Lasso</b>를 사용하기로 결정

```python
from sklearn.linear_model import Ridge, Lasso

# Ridge
ridge = Ridge(alpha=1e-4, random_state=64)

# Lasso
lasso = Lasso(alpha=1e-3, random_state=64)

ridge_model = ridge.fit(X_train_std, y_train)
lasso_model = lasso.fit(X_train_std, y_train)
```

```python
print('Ridge:', ridge_model.score(X_train_std, y_train), ridge_model.score(X_test_std, y_test))
```

```python
print('Lasso:', lasso_model.score(X_train_std, y_train), lasso_model.score(X_test_std, y_test))
```

### 예측값과 실제값 비교 그래프(Ridge)

```python
mpl.rc('font', family='NanumGothic')  # matplotlib에서 한글 폰트로 NanumGothic을 사용
mpl.rcParams['axes.unicode_minus'] = False

w_ = 0.3
plt.figure(figsize=(20, 10))

pred = ridge_model.predict(X_test[:])
plt.bar(range(len(pred)), pred, width=w_)

true = y_test.values[:]
plt.bar([i + w_ for i in range(len(true))], true, width=w_)

plt.ylabel('승률')
plt.legend(('예측 승률', '실제 승률'))

plt.show()
```

## 딥러닝

샘플 개수가 적을 수록 과대적합이 쉽게 일어나기 때문에 작은 모델을 사용하는 것이 과대적합을 피하는 방법이 될 수 있다.

본 프로젝트에서는 64개의 유닛을 가진 2개의 은닉 층으로 작은 인공신경망을 구성해서 사용한다.

### 사용할 데이터

다중공선성 검사를 통해 선택한 Feature로 모델 생성

### 모델 구성

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense

def build_model():
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

### 모델 검증
K-fold cross-validation(K-겹 교차 검증) 사용

```python
X_train.head()
```

```python
k = 4

num_val_samples = len(X_train) // k
num_epochs = 300
all_mae_histories = []
for i in range(k):
    print('fold #' + str(i))
    
    # 검증 데이터 준비; k번째 분할
    X_val = X_train[i * num_val_samples: (i + 1) * num_val_samples]
    y_val = y_train[i * num_val_samples: (i + 1) * num_val_samples]
    
    partial_X_train = np.concatenate(
        [X_train[:i * num_val_samples],
         X_train[(i + 1) * num_val_samples:]],
        axis=0)
    
    partial_y_train = np.concatenate(
        [y_train[:i * num_val_samples],
         y_train[(i + 1) * num_val_samples:]],
        axis=0)

    model = build_model()
    history = model.fit(partial_X_train, partial_y_train, validation_data=(X_val, y_val), epochs=num_epochs, batch_size=1, verbose=0)  # verbose: 훈련 과정 표시 여부
    mae_history = history.history['val_mean_absolute_error']
    all_mae_histories.append(mae_history)
```

```python
all_mae_histories
```

```python
average_mae_history = [np.mean([x[i] for x in all_mae_histories]) for i in range(num_epochs)]
```

```python
plt.plot(range(1, len(average_mae_history) + 1), average_mae_history)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

```python
test_mse_score, test_mae_score = model.evaluate(X_test, y_test)
print(test_mse_score, test_mae_score)
```

```python
mpl.rc('font', family='NanumGothic')  # matplotlib에서 한글 폰트로 NanumGothic을 사용
mpl.rcParams['axes.unicode_minus'] = False

w_ = 0.3
plt.figure(figsize=(20, 10))

pred = model.predict(X_test[:])

pred = pred.reshape(len(pred),)
plt.bar(range(len(pred)), pred, width=w_)

true = y_test.values[:]
plt.bar([i + w_ for i in range(len(true))], true, width=w_)

plt.ylabel('승률')
plt.legend(('예측 승률', '실제 승률'))

plt.show()
```
